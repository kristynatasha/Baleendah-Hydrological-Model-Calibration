# 1 <= CH <= 5 mm

```python
CH = np.zeros((42,54,60))
probability = np.zeros((42,54,60))

for time in np.arange(33,42):
    for i in range(0,59,1):
        for j in range(0,53,1):
            CH[time,j,i]=0
            for k in range(0,9,1):
                if (ens[k,time,j,i]>=1.0) & (ens[k,time,j,i]<=5.0):
                    CH[time,j,i] = CH[time,j,i] + 1
probability = CH/9
print(probability.shape)
```

```python
import matplotlib as mpl
levels = np.array([0.0,0.125,0.25,0.375,0.5,0.625,0.75,0.875,1.0])

for xx in np.arange (33,42,3):
    rain = getvar(data1, "RAINC",xx)
    lats, lons = latlon_coords(rain)
    bm = get_basemap(rain,resolution='h')
    
    fig = plt.figure(figsize=(12,9))
    
    parallels = np.arange(-1,5, 0.6)
    meridians = np.arange(96,102, 0.6)

    bm.drawparallels(parallels,labels=[1,0,0,0],fontsize=12,color='black')
    bm.drawmeridians(meridians,labels=[0,0,0,1],fontsize=12,color='black')
    
    bm.drawcoastlines(linewidth=2)
    bm.drawstates(linewidth=2)
    bm.drawcountries(linewidth=2)
    
    x, y = bm(to_np(lons), to_np(lats))
    
    cmap =mpl.colors.ListedColormap(['white','dodgerblue', 'blue','limegreen','yellow','orange','red','maroon'])

    cs = bm.contourf(x,y,to_np(probability[xx,:,:]),levels,cmap=cmap)
    cbar = bm.colorbar(cs,location='right', shrink=.62,pad="5%")
    cbar.set_label('Probability',fontsize=12)

    plt.title("Probability Maps Presipitasi >= 1 mm & <= 5 mm - T" + str(xx), fontsize=15)

    plt.show()

    fig.savefig('Prob_CH_1mm'+str(xx) +'.png')
```
