```python
    Temp1 = getvar(file1,"temp",units='degC',timeidx=30)
    Hgh1= getvar(file1,"p",units='hPa',timeidx=30)
    ting1=getvar(file1,"z",timeidx=30)
    TD=getvar(file1, 'td', timeidx=30)
    Tempv1 = (Temp1[:,25,26])
    Hghv1=(Hgh1[:,25,26])
    Tingv1=ting1[:,25,26]
    TD1=TD[:,25,26]

    Temp2 = getvar(file2,"temp",units='degC',timeidx=30)
    Hgh2= getvar(file2,"p",units='hPa',timeidx=30)
    ting2=getvar(file2,"z",timeidx=30)
    TD=getvar(file2, 'td', timeidx=30)
    Tempv2 = (Temp2[:,25,26])
    Hghv2=(Hgh2[:,25,26])
    Tingv2=ting2[:,25,26]
    TD2=TD[:,25,26]
    
    Temp3 = getvar(file3,"temp",units='degC',timeidx=30)
    Hgh3= getvar(file3,"p",units='hPa',timeidx=30)
    ting3=getvar(file3,"z",timeidx=30)
    TD=getvar(file3, 'td', timeidx=30)
    Tempv3 = (Temp3[:,25,26])
    Hghv3=(Hgh3[:,25,26])
    Tingv3=ting3[:,25,26]
    TD3=TD[:,25,26]
```

```python
# Calculate the LCL
lcl_pressure, lcl_temperature = mpcalc.lcl(Hghv1[0], Tempv1[0], TD1[0])

print(lcl_pressure, lcl_temperature)

# Calculate the parcel profile.
parcel_prof = mpcalc.parcel_profile(Hghv1, Tempv1[0], TD1[0])

# Create a new figure. The dimensions here give a good aspect ratio
fig = plt.figure(figsize=(9, 9))
skew = SkewT(fig, rotation=30)

# Plot the data using normal plotting functions, in this case using
# log scaling in Y, as dictated by the typical meteorological plot
skew.plot(Hghv1, Tempv1, 'r')
skew.plot(Hghv1, TD1, 'g')
skew.ax.set_ylim(1000, 100)
skew.ax.set_xlim(-40, 60)

# Plot LCL temperature as black dot
skew.plot(lcl_pressure, lcl_temperature, 'ko', markerfacecolor='black')

# Plot the parcel profile as a black line
skew.plot(Hghv1, parcel_prof, 'k', linewidth=2)

# Plot a zero degree isotherm
skew.ax.axvline(0, color='c', linestyle='--', linewidth=2)

# Add the relevant special lines
skew.plot_dry_adiabats()
skew.plot_moist_adiabats()
skew.plot_mixing_lines()

# Show the plot
plt.title('Skew-T Sidoarjo 17 Januari 2021 13.00 WIB dengan Skema YSU', fontsize ='14')
plt.show()
```
